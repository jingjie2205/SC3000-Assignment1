For task 2, implementing a uninformed search algorithm, we will implement UCS (Uniform Cost Search) Due to the constraint of having a total energy cost.

Comparing the functionality of UCS to Dijkstra, Dijskstra finds the shortest path from the start node to all other nodes in the graph, whereas UCS finds the path with the lowest cost to the goal node, making UCS more efficient in time complexity than Dijkstra

```python
import heapq
import json

def load_graph_and_cost():
    G_file = "../data/G.json"
    Cost_file = "../data/Cost.json"

    with open(G_file, 'r') as f:
        G = json.load(f)
    with open(Cost_file, 'r') as f:
        Cost = json.load(f)
    return G, Cost

def uniform_cost_search_with_energy(G, Cost, start, goal, max_energy):
    pq = [(0, start, [start])] 
    visited = set()
    
    while pq:
        cost, node, path = heapq.heappop(pq)
        
        # Fail if cost exceeds energy
        if cost > max_energy:
            continue
        
        if node == goal:
            return path, cost
        
        if node in visited:
            continue
        visited.add(node)
        
        for neighbor in G.get(node, []):
            if neighbor not in visited:
                # Get the cost of edge from node to neighbor, e.g "1,2". If we just put node / neighbour we dont know the node to travel to
                edge_key = f"{node},{neighbor}"
                edge_cost = Cost.get(edge_key)
                new_cost = cost + edge_cost
                if new_cost <= max_energy:
                    heapq.heappush(pq, (new_cost, neighbor, path + [neighbor]))
    
    # No path found within energy limit
    return None, None
```

```python
if __name__ == "__main__":
    G, Cost = load_graph_and_cost()

    path, cost = uniform_cost_search_with_energy(G, Cost, "1", "50", 287932)
    if path:
        print(f"Path: {path}, \nCost: {cost}")
        print(f"Length of path: {len(path)}")
    else:
        print(f"No path found within energy limit of 287932 energy.")
```
